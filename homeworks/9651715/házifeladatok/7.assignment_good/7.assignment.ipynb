```python
import os 
import sys
import time

sys.path.append("../../../..")

from src.pricer import *
MarketData.initialize()
```

```python
und = Stock.TEST_COMPANY
ls = LongShort.LONG
strike = 1.2
expiry = 1.0
strike_level = strike * MarketData.get_spot()[und]
model = FlatVolModel(und)

contract = EuropeanContract(und, PutCallFwd.CALL, ls, strike_level, expiry)

mc_params = MCParams(num_of_path=20000, tenor_frequency=12, seed=1, evolve_spot_method=MCNumMethod.MILSTEIN)
pricer_mc = GenericMCPricer(contract, model, mc_params)

start_time_old = time.time()

delta_old = Pricer.calc_delta(pricer_mc, GreekMethod.BUMP) 
end_time_old = time.time()

time_old = end_time_old - start_time_old


start_time_new = time.time()

delta_new = pricer_mc.calc_delta(GreekMethod.BUMP)
end_time_new = time.time()

time_new = end_time_new - start_time_new



print("\n[ EREDMÉNYEK ]")
print(f"RÉGI módszer Delta: {delta_old:.6f} | Futási idő: {time_old:.4f} másodperc")
print(f"ÚJ módszer Delta:   {delta_new:.6f} | Futási idő: {time_new:.4f} másodperc")
print("-" * 65)

speedup = time_old / time_new
print(f"az ÚJ módszer {speedup:.1f}-szer gyorsabb!")
```
